```python
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import requests
from io import StringIO


file_url = 'https://drive.google.com/uc?export=download&id=1ht6FoMfneRyFo2Iv8mvevjmZFmDJffTi'
response = requests.get(file_url)
data = pd.read_csv(StringIO(response.text))


sns.regplot(data=data, x='sat_sum', y='fy_gpa', scatter_kws={'alpha': 0.5})
plt.title('Relationship between GPA and SAT Scores')
plt.xlabel('SAT Summary (math + verbal)')  
plt.ylabel('GPA')
plt.show()



X = data['sat_sum']
y = data['fy_gpa']
X = sm.add_constant(X)  
model = sm.OLS(y, X).fit()

summary = model.summary()
summary_text = summary.tables[0].as_text() + "\n" + summary.tables[1].as_text()
summary_text = summary_text.replace("Warnings:\n", "")  
print(summary_text)



```
